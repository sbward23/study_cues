# Alcohol and sexual consent communication

A daily diary investigation of verbal and nonverbal cues

## Introduction

Sexual consent is a complex process involving thoughts, feelings, and behaviors used to determine and express willingness to engage in sexual activity.<sup>[1](#ref-muehlenhardComplexitiesSexualConsent2016)</sup> Specific behaviors used to infer and communicate consent include verbal and nonverbal cues, which can be explicit or implicit, ranging from clear spoken dialogue (“Yes!”) to subtle body language.<sup>[2](#ref-willisAssociationsInternalExternal2019)</sup> The presence of consent distinguishes wanted sexual experiences from sexual assault, so programs aiming to prevent sexual assault often target consent communication. A substantial gap exists between these programs’ curricula and the varied consent practices that young people use during sexual encounters, particularly those involving alcohol use.<sup>[3](#ref-curtis2017)</sup>

Alcohol use frequently co-occurs with sexual assault and muddles communication dynamics within young adults’ consensual sexual encounters. A recent scoping review of the alcohol and sexual consent literature highlighted the need for more quantitative research in this area. The authors called for studies using methods that permit careful evaluation of the contextual features of alcohol-involved consent by facilitating event-level data collection on individual sexual encounters and aspects within them that covary with cue use. This call to action is important because contextual factors—such as co-occurring intoxication and type of sexual activity—have been linked to within-person differences in consent communication.<sup>[5](#ref-willisInternalExternalSexual2021)</sup>

Researchers studying contextual factors linking alcohol to variance in consent communication have done so by conducting systematic, within-person comparisons of two matched events,<sup>[6](#ref-marcantonioWithinPersonExaminationAlcoholInvolved2024)</sup> or by grounding responses to each person’s most recent sexual encounter, which could have occurred weeks to months in the past. While informative, these studies risk greater recall bias by asking people to remember past events. In contrast, micro-longitudinal studies assess contextual factors closer in time to when consent communication occurs. Although ecological momentary assessment and electronic daily diaries have been used extensively in other clinical science research to illuminate important time-varying predictors of health risk behaviors,<sup>[9](#ref-bolger2003)</sup> only one published investigation of this kind has been examined alcohol and sexual consent communication.

Willis and colleagues surveyed participants three times per day for 28 days and found that within-person variability accounted for more than 70% of the variance in sexual consent communication.<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup> Participants reported greater consent cue use for sexual encounters involving more physically intimate sexual behaviors (i.e., vaginal or anal penetration) than for encounters involving less intimate behaviors,<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup> but no significant event-level associations between alcohol involvement and verbal or nonverbal consent were observed in this sample.<sup>[5](#ref-willisInternalExternalSexual2021)</sup> The previous research examined a limited set of predictors and reflected a specific subpopulation of adults in committed relationships; the current project builds on their findings by probing encounter- and person-level associations with alcohol and consent cue use among university students while expanding the range of possible predictors. Examining these associations in different samples and in relation to a wider array of variables is a crucial next step toward isolating more actionable targets for sexual assault prevention efforts focused on changing consent behavior.

### Between-person predictors of active consent cues

Rates of sexual assault on U.S. college campuses remain virtually unchanged after more than 30 years of dedicated research and intervention.<sup>[12](#ref-kossScopeRapeVictimization2022)</sup> Lack of consent is increasingly recognized as defining characteristic of sexual assault, leading institutions to incorporate consent-focused education into violence prevention efforts.<sup>[14](#ref-beresRethinkingConceptConsent2014)</sup> University policy and educational initiatives promote explicit and enthusiastic consent communication (“only yes means yes!”), while young people often rely instead on passive cues (not resisting, not saying no) and assumptions (“We love each other”)<sup>[15](#ref-willisSexualConsentCommitted2021)</sup> to determine consent.<sup>[17](#ref-jozkowskiGenderDifferencesHeterosexual2014)</sup> Sexual consent education aims to promote the use of active cues by enhancing consent knowledge and positive attitudes while decreasing rape-supportive attitudes, presuming that attitudes shape consent behavior. However, these interventions do not have a clear underlying mechanism of behavior change, and they rarely assess behavioral outcomes.<sup>[18](#ref-burtonTeachingSexualConsent2023)</sup> A prerequisite to improving upon the value of consent education is testing whether these attitudinal intervention targets predict actual in-the-moment cue use, and contrasting their explanatory power with that of other predictors.

### Attitudes and adherence to social norms

#### Individual differences

Rape-supportive attitudes and beliefs have been theorized to explain variability in how individuals approach consent communication.<sup>[19](#ref-jozkowskiChapter9Sexual2022)</sup> Perhaps the most prominent example in the sexual assault and consent literatures concerns Rape Myth Acceptance (RMA), the endorsement of widely-held false beliefs that justify sexual aggression and place blame on victims.<sup>[21](#ref-payneRapeMythAcceptance1999)</sup> RMA is a robust predictor of sexual violence perpetration; it also has been used as an outcome measure in several sexual consent interventions delivered to young people in educational settings.<sup>[22](#ref-baker2014)</sup> Individuals with high levels of RMA often hold the belief that victims of sexual assault “led the other person on” (e.g., with what could be conceived through this lens as nonverbal consent cues) or provided verbal agreement and “later regretted it.” These pervasive schemas suggest individuals with high levels of RMA may pursue sexual contact without first communicating their willingness or looking for indications of the other person’s openness to their advances.<sup>[24](#ref-mcmahonUpdatedMeasureAssessing2011)</sup> There also are rape myths that specifically excuse perpetrator behavior in the context of alcohol intoxication (e.g., “It shouldn’t be considered rape if the person is drunk and didn’t know what they were doing”). Rape myths are distinct from but related to permissive Alcohol and Sexual Consent (ASC) attitudes,<sup>[24](#ref-mcmahonUpdatedMeasureAssessing2011)</sup> both of which contradict sexual assault programming messages about, for example, how communication signals are easily misinterpreted when alcohol is involved in a sexual situation. Permissive ASC attitudes reflect social norms that are prevalent on college campuses (e.g., “Consensual drunk sex is a normal and harmless part of college life”) and reinforce approval of intoxicated sexual activity without regard for the associated risks. Our previous research found that permissive ASC attitudes were negatively related to ongoing and communicative (i.e., “process-based”) consent behavior among college men.<sup>[26](#ref-wardAlcoholSexualConsent2012)</sup>

#### Perceived-peer differences

Among college men, peer relationships can be critical to their risk of perpetrating sexual aggression<sup>[28](#ref-swartoutTrajectoryAnalysisCampus2015)</sup> and male peer support for abuse and coercion have been linked to lower comprehension of sexual consent and less communicative sexuality.<sup>[29](#ref-warrenComprehensionSexualConsent2015)</sup> The relevance of a person’s beliefs about their peers’ ideas of what is normal and acceptable is best contextualized by the “sociality of consent,” which refers to a phenomenon wherein sexual consent is not determined privately between partners but rather extended into the peer group.<sup>[30](#ref-jensen2020)</sup> Qualitative research assessing how young adults use relational and social contexts to determine consent further underscores the salience of peer influence, as participants–particularly college women–described their friends asking them to help interpret and even make judgments about consent on their behalf after consuming alcohol or other drugs.<sup>[33](#ref-marcantonioWhatWouldYou2025)</sup>

### Typical and historical behavior

#### Hazardous alcohol use and sexual risk-taking

Alcohol is a salient factor in the context of both sexual assault and consensual sex among college students, yet sexual assault prevention and risk-reduction programs rarely include information about how alcohol can affect consent communication.<sup>[17](#ref-jozkowskiGenderDifferencesHeterosexual2014)</sup> While scholars and practitioners have started integrating alcohol and sexual assault programming,<sup>[35](#ref-gilmore2025)</sup> there is far more research on alcohol-related sexual assault compared to consent involving alcohol use. Some research suggests that young adults who engage in binge drinking generally use less active consent communication than those who do not binge drink,<sup>[16](#ref-marcantonioEffectsTypicalBinge2022)</sup> but the intensity of alcohol consumption at the encounter level may be more directly relevant to in-the-moment. Still, cross-sectional surveys cannot establish whether real-world drinking events decrease the incidence of active consent cues, and few micro-longitudinal studies have explicitly investigated situational factors linked to within-person differences in consent communication.

### Situational correlates of consent cue use

#### Event-level alcohol consumption

As previously mentioned, Willis and colleagues found no significant event-level associations between alcohol use and consent cues.<sup>[5](#ref-willisInternalExternalSexual2021)</sup> A daily diary dissertation study found that, among college women engaging in casual sex, Blood Alcohol Estimate was negatively associated with consent expression, but this relationship was no longer significant when the sexual behavior type was included in the model.<sup>[4](#ref-rinioloRoleAlcoholContext2025)</sup> One potential explanation is that an individual’s alcohol consumption is more likely to result in them communicating consent differently than they typically would on days when their alcohol use is also atypical–representing a deviation from their own personal norm. This hypothesis is supported by previous research which shows that, on days when women drink more than their own average, they are more likely to experience negative sexual consequences later that day.<sup>[37](#ref-scaglione2014)</sup>

#### Event-level perceived intoxication

Young adults also recognize and describe themselves adjusting the strategies they use to communicate sexual consent based on their perceptions of their own intoxication relative to that of a potential partner, but this recurring theme in qualitative and vignette-based research has proven difficult to explicate quantitatively.<sup>[38](#ref-laughlinJustOneShot2023)</sup> One hypothesis is that the extent to which perceived differences between one’s own and their partner’s intoxication relate to their use of verbal consent cues depends on their level of familiarity with that partner. For example, if someone knows the other person involved in a sexual encounter well and is motivated to maintain a positive relationship with them moving forward, they may be more motivated to voice their own consent verbally and to check in with that other person about their degree of intoxication and capacity to consent before proceeding with sexual activity. This hypothesis is supported by the ‘date rape’ literature, which indicates that sexual assaults involving mutual alcohol consumption are more likely to take place outside the home and involve someone less well known to the victim.<sup>[39](#ref-lorenzAlcoholSexualAssault2016)</sup>

#### Partner familiarity and sexual precedent

The relationship or degree of familiarity between two people involved in a sexual encounter has been related to differences in consent communication, though existing research has produced paradoxical conclusions regarding the nature of these associations. Sexual precedence–the idea that engaging in consensual sex at one point in time implies consent can be assumed for future encounters with that same person–is supported by research showing people perceive a greater need to directly communicate about consent when partners are not in a committed relationship.<sup>[40](#ref-humphreysPerceptionsSexualConsent2007)</sup> Yet, cross-sectional survey research among U.S. college students indicates that explicit verbal consent cues are less likely to be used when the sexual relationship is casual. Internal consent feelings (safety, comfort) tend to be higher for encounters with committed dating partners and lower for one-time partners,<sup>[41](#ref-walshDualMeasuresSexual2019)</sup> and internal consent feelings are strong predictors of consent cue use, but patterns observed over time within relationships suggest a complex relationship between shared sexual history and consent communication that warrants further evaluation.<sup>[42](#ref-willisSexualPrecedentsEffect2019)</sup>

### Significance of the current research

Conflicting findings and gaps in the literature highlight a pressing need for research grounded in real-time lived experiences that can help to explain what relational and situational factors are linked to differences in the degree to which university students’ use different consent cues. Experiential sampling via daily reports can facilitate a more nuanced understanding of the types of alcohol involvement that weaken verbal and nonverbal consent communication during students’ sexual encounters. Knowledge of these contextual factors can then be used to inform ecological momentary interventions designed to be delivered when and where the evidence suggests they will be most effective–an approach that has shown promise for reducing risk of sexual revictimization among college women.<sup>[43](#ref-maloney2024)</sup>

### Research aims and hypotheses

Before addressing primary study aims, our preliminary goal was to evaluate relationships between sexual event contextual characteristics and cue use for communicating about consent during sexual encounters. Gender, for instance, is a significant factor,<sup>[44](#ref-willisBarriersSuccessAffirmative2018)</sup> but there is inconsistent evidence regarding its associations with consent cues. Some research suggests men are less likely to use nonverbal cues and more likely to use verbal cues when compared to women or queer individuals, yet other studies observe interactions or trends that imply gender differences hinge on context.<sup>[15](#ref-willisSexualConsentCommitted2021)</sup> Previous research also has identified relational factors with theorized links to alcohol and sexual consent communication,<sup>[45](#ref-marcantonio2025)</sup> but results characterizing the magnitude and direction of these associations has been mixed. Unpacking complexity around these person-level and situational factors could yield valuable information about the conditions under which particular individuals are more or less likely to use certain types of consent cues.

**We tested the following preliminary hypotheses to clarify whether sexual event contextual factors covary with cue use in the current sample.**

-   H0a: Verbal consent cue use will be greater for encounters involving penetrative sexual behaviors when compared to encounters that did not advance beyond kissing/touching.

-   H0b: Nonverbal consent cue use will be greater for encounters involving more familiar partners when compared to encounters involving partners with less familiarity.

-   H0c: Nonverbal consent cue use will be greater for women and gender nonconforming (GNC) individuals when compared to men.

**Aim 1: To examine associations between situational alcohol involvement and consent cue use during sexual encounters.**

-   H1a: Alcohol consumption (# of drinks) will be significantly, negatively associated with both verbal and nonverbal consent cue use.

-   H1b: The continuous absolute difference between subjective participant and perceived partner intoxication (“intoxication disparity”) will be significantly, negatively associated with verbal cue use.

-   H1c: Drinking venue attendance will be significantly, negatively associated with both nonverbal and verbal cue use. We expect this negative effect will be more pronounced for nonverbal cues.

-   H1d: Among alcohol-involved encounters, there will be an inverted-U relationship between both alcohol consumption and subjective intoxication and verbal cue use: low → moderate = verbal consent increases; moderate → high = verbal consent decreases.

**Aim 2: To investigate the moderating role of sexual event contextual factors in the associations between alcohol involvement and consent cue use.**

-   H2a: The negative association between alcohol consumption and verbal cue use will be weaker for penetrative behaviors and stronger for encounters involving kissing/touching.

-   H2b: The negative association between drinking venue attendance and nonverbal cue use will be stronger among less familiar partners.

-   H2c: The negative association between intoxication disparity and verbal cue use will be stronger among men compared to women and GNC individuals.

**Aim 3: To evaluate the extent to which distal individual factors predict consent cue use after controlling for partner familiarity and sexual behavior.**

-   H3a: Social desirability bias and rape myth rejection will be positively associated with verbal cues after controlling for partner familiarity and behavior type.

-   H3b: Permissive individual and perceived peer attitudes toward alcohol and coercion will be negatively associated with verbal cues after controlling for partner familiarity and behavior type.

-   H3c: Hazardous alcohol consumption will be negatively associated with both verbal and nonverbal cues after controlling for partner familiarity and behavior type.

-   H3d: Permissive attitudes toward alcohol and coercion will strengthen the negative association between intoxication disparity and verbal cue use after controlling for partner familiarity and behavior type.

## Method

The current research used baseline and 28-day daily diary data to move beyond the limitations of prior studies, allowing us to consider different forms of alcohol involvement in relation to sexual event context, individual difference factors, and verbal and nonverbal sexual consent cue use. Any distal, person-level predictors identified as important will have their relative contributions verified in a model that also includes proximal situational factors, increasing confidence that these are reliable targets for intervention or further theory development. Our lab collected these data from undergraduates at a large public university in the midwestern U.S. during the Fall 2022 semester.

### Participants

We used Sona Systems to recruit participants from an online subject pool of undergraduates taking introductory psychology courses. We invited eligible students to participate via email. Participants were made aware that the study had two parts: a 30-minute baseline survey, and up to 28 daily diary surveys. At the end of the baseline survey, students were asked whether they had a smartphone AND whether they wanted to opt into the daily diary portion of the study. Those who responded affirmatively to both were enrolled in and pushed daily survey links through SMS for 28 days. 

Eligible participants were those 18 years or older who owned a smartphone and reported consuming at least one alcoholic beverage in the past year. A total of 386 students met eligibility criteria and completed at least 20% of the baseline assessment. Of these, 278 students participated in the daily diary portion of the study. Participant ages ranged from 18 - 23.

### Procedures

The research team used Qualtrics to construct a series of data collection instruments for web-administration, and SurveySignal to send daily text messages with links to each daily diary survey. After providing informed consent, participants gained access to the baseline assessment via Qualtrics. After completing the baseline, each participant received (1.0) class credit as compensation, which translates to an additional percentage point on their final grade in their introductory psychology course. Students with a smartphone were then given the option to participate in the daily diary portion of the study. Every day for 28 days, each participant who opted in received a text message between 11:00 AM and 2:00 PM with a link to that day’s survey. Each day, those who had not yet opened the survey link also received a reminder text message two hours after the initial text was sent. For each complete daily diary, participants received 0.25 class credits as compensation. To incentivize responding, participants who completed all 28 daily diary surveys received a \$20 Amazon.com gift card in addition to the 7.0 extra credit percentage points (28 x 0.25 class credits).

The baseline and daily diary surveys were designed to capture theoretically and empirically supported personal and situational factors, respectively. The measures we selected were drawn from peer-reviewed research examining sexual consent and assault.

## Measures

### Baseline

Demographic measures assessed gender, age, race, sexual orientation, year in school, relationship status, campus involvement (e.g., Greek life, athletics), residence (e.g., dormitories), disability status, and socioeconomic status indicators (e.g., Pell grant recipient). Rape myth rejection was assessed with the Illinois Rape Myths Acceptance scale<sup>[24](#ref-mcmahonUpdatedMeasureAssessing2011)</sup>. Items from the Alcohol and Sexual Consent scale,<sup>[26](#ref-wardAlcoholSexualConsent2012)</sup> the Process-Based Consent Scale (subtle coercion sub-scale),<sup>[46](#ref-glaceMovingNewModel2020)</sup> and the Rape Attitudes and Beliefs Scale<sup>[47](#ref-burgess2007)</sup> were used to assess individual and perceived peer attitudes toward alcohol and sexual coercion. Hazardous alcohol use was assessed with the Alcohol Use Disorders Identification Test<sup>[48](#ref-saundersDevelopmentAlcoholUse1993)</sup> and typical consent behavior was assessed with the Sexual Consent Scale-Revised.<sup>[49](#ref-humphreysSexualConsentScalerevised2010)</sup>

### Daily Diary

Each daily diary prompted participants to report on behaviors and experiences since their last survey and began with five behavioral screening items assessing (1) partnered sexual behavior, (2) alcohol use behavior, (3) online sexual behavior, (4) solitary sexual behavior, and (5) drug use behavior. We designed the daily surveys so that each participant would complete 1-2 survey blocks per day assigned based on their responses to the screening items. When a participant responded “No” or “Prefer not to respond” to all five screening items with corresponding behavior blocks, they were assigned to a general mental health block. Survey blocks were all comparable in length and focused on sensitive topics to discourage selective responding and ensure that each person received roughly the same number of follow-up questions per day, regardless of the experiences they endorsed on the screening items.

Survey blocks were rank-ordered from 1-5 and prioritized partnered sexual behavior and alcohol use behavior data, such that, whenever a participant responded “Yes” to more than two screening items, branching logic directed them to the two-highest ranked survey blocks. The screening item for partnered sexual behavior read, “Since the last survey, have you hooked up, engaged in any sexual activity, or tried to engage in any sexual activity with another person (including flirting, kissing, touching, or intercourse)?” Response options were: Yes (= 1), No (= 0), Prefer not to respond (= 3). 

#### Sexual event context

**Partner familiarity:** Participants responded to the question, “How did you know this other person?” by selecting one of the following familiarity levels from a list of response options: “I didn’t, we just met that day;” “They are an acquaintance;” “They are a friend;” and “They are my current or past dating/intimate partner.” We collapsed the ‘acquaintance’ and ‘friend’ response categories into one level, which was compared against dating/intimate partners and people who just met via factor variable `fam_lev`.

**Sexual precedent:** Participants responded to the question, “Had you ever had sexual contact with this person in the past?” with response options Yes (=1) and No (=0). 

**Sexual behavior type:** Participants were instructed to select all that apply from a list of specific behaviors that captures escalating levels of physical intimacy aligned with previously established sexual hierarchies: Kissing (=1); Touching (=2); Penetration with fingers or fingering (=3); Oral sex (=4); Vaginal sex (=5); and Anal sex (=6).<sup>[50](#ref-sanders2010)</sup> An additional response option read, “I tried but no sexual contact happened,” and encounters for which this was the only response will not be included in the present analysis. A dichotomous variable comparing events involving Contact (kissing, touching) with those involving Penetration was created; the Contact v. Penetration distinction is what differentiates Sexual Assault from Rape in policy and criminal justice contexts as well as within the scientific literature. Limited representation across sexual behavior types precluded the use of an ordinal event-level variable during hypothesis testing.

#### Alcohol involvement

**Perceived intoxication:** Participants rated their own intoxication (“How intoxicated did you feel right before or during this sexual activity?”) and their perception of their partner’s intoxication (“How intoxicated do you think the other person felt right before or during this sexual activity?”) using a sliding scale from “Least intoxicated ever felt” (=0) to “Most most intoxicated ever felt” (=100). 

**Intoxication disparity:** A continuous intoxication “disparity” variable was calculated as the absolute difference between self-reported intoxication and perceived partner intoxication. Larger values indicate lower perceived alignment between self- and partner- intoxication. We also created a factor variable with categories for equal intoxication (Parity), for when the participant perceived themselves as more intoxicated (Self-Higher), and for when the participant perceived the other person as more intoxicated (Partner-Higher). 

**Pre-encounter alcohol consumption:** Participants were asked, “How many drinks did you consume prior to or during this activity?” with response options for the following drink counts: 0; 1-2; 3-4; 5-6; 7-9; and 10+ drinks. A continuous `drink` variable was created from this item, ranging from 0-5.

**Drinking venue attendance:** Participants were asked, “Since the last survey, have you been to a bar, club, or party?” with response options for “Yes” (=1) and “No” (=0). This item was included in the first survey block and presented to participants after they completed the five behavioral screeners, but before branching logic routed them to their behavior-specific blocks for that daily report.

#### Validity of outcome measures

Despite the need for and potential benefits of using experience sampling methods (ESM) to examine alcohol-involved sexual consent experiences, there also are unique challenges associated with this approach. One such challenge is the increased commitment and investment of effort that repeatedly completing assessments requires from participants.<sup>[52](#ref-vanhooff2007)</sup> ESM researchers can mitigate participant burden by making surveys brief, a common but imperfect solution that frequently necessitates measuring variables with one or just a few items per construct.<sup>[53](#ref-fisher2012)</sup> Measures composed of only one or two items are often criticized by researchers with well-founded concerns that they might be unreliable, more susceptible to measurement error, or fail to capture complex constructs.<sup>[54](#ref-robins2001)</sup> Still, other researchers argue that relying on single-item measures is acceptable when these items are carefully developed and evaluated for multiple sources of measurement validity.<sup>[55](#ref-wolfers)</sup>

Recent recommendations from the field of communication science propose three central criteria for validating single-item measures in an ESM context.

The item must:

1.  Fit the core of a construct’s definition to ensure that it reflects the construct well.
2.  Be sensitive to situational changes that demonstrate predictable within-person variability
3.  Be sensitive to differences between groups at the between-person level

In the sections that follow, I present findings aligned with these criteria that offer support for the validity of the brief outcome measures used in the current research.

##### 1) Definitional correspondence

Active consent communication is the use of words or behaviors (i.e., cues) to communicate to another person agreement to engage in sexual activity. It is also referred to as “external” consent communication because these words and behaviors are used to outwardly express one’s internal feelings of willingness to participate in the sexual behavior. The only validated external consent measure that existed until recently was the 18-item External Consent Scale, which was developed for cross-sectional research to ask about the respondent’s most recent sexual encounter (which could have been days to weeks in the past).<sup>[41](#ref-walshDualMeasuresSexual2019)</sup> Because including 18 items in ESM studies would be time-intensive and burdensome, Willis and colleagues used a two-phase procedure to develop and evaluate brief measures of external consent cues specifically designed for use in ESM studies.<sup>[51](#ref-willisDevelopingValidFeasible2021)</sup>

In Phase I, the research team sought to develop items that best represented the constructs of interest while demonstrating both face validity and content validity. To do so, operational definitions for the external and internal consent constructs were derived from seminal theoretical research and systematic reviews of the sexual consent literature. Cognitive interviews were then conducted with sexually active adults (n = 10) using concurrent probing to assess sample item comprehension, preferred wording, and interpretation. Three content-matter experts from the field of sexual consent research and three measurement experts with doctoral training in psychometrics also provided ratings to determine the degree to which sample items captured their corresponding operational definitions. The research team then calculated item-objective congruence (IIOC) values for content alignment (cutoff .75) to establish which prospective items had higher content validity.<sup>[57](#ref-turner2003)</sup> Items were selected for Phase II by balancing participants’ judgments and experts’ IIIOC ratings.<sup>[51](#ref-willisDevelopingValidFeasible2021)</sup>

The brief measure of external consent<sup>[51](#ref-willisDevelopingValidFeasible2021)</sup> consists of four items that were developed to measure four types of active consent cues identified through previous research: verbal (“I verbally communicated…”), nonverbal (“I nonverbally communicated…”), explicit (i.e., direct; “I used straightforward signals to communicate…”) and implicit (ie., indirect; “I used subtle signals to communicate…”). Researchers using the measure have tried to operationalize active external consent using a composite score derived from participants’ responses to these four items, with variable success. Across three studies by Shi and colleagues using three separate datasets, their team found that the four-item composites consistently had poor reliability (αs \< .60).<sup>[58](#ref-shiImpactSexualConsent2025)</sup> This pattern of findings aligned with that of another study by Willis and colleagues, wherein internal reliability was not strong for a composite including verbal, nonverbal, explicit, and implicit consent cues.<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup> Moreover, across the Shi<sup>[58](#ref-shiImpactSexualConsent2025)</sup> and Willis<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup> studies, verbal cues were highly positively correlated with explicit cues, and nonverbal cues were highly positively correlated with implicit cues.

In the pilot study that was initially conducted to validate the external consent measure, the research team solicited open-ended feedback from participants. When queried about whether any of the survey questions sounded awkward or lacked clarity, a few participants expressed confusion regarding the explicit item’s phrasing (e.g., “It is unclear to me what this question (on straightforward signals) means”), as well as uncertainty about the distinction between ‘nonverbal’ and ‘subtle’ communication (e.g., “how is nonverbal different than ‘subtle’ communication?”, “What differentiates the two?”).<sup>[51](#ref-willisDevelopingValidFeasible2021)</sup> Considering pilot participant feedback, strong positive correlations between verbal and explicit cues reported in previous studies,<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup> and the goals of the current research, we decided against including the explicit/“straightforward” cue item in our daily diary. We did include both the nonverbal and implicit cue items, but we observed the same pattern described in previous research. Specifically, nonverbal cues were highly positively correlated with implicit cues (r = .73). Considering this strong correlation and the preregistered strategy for addressing it reported in prior work,<sup>[58](#ref-shiImpactSexualConsent2025)</sup> the nonverbal and implicit cue items were averaged to create a composite Nonverbal score. Finally, the internal consistency for the verbal, nonverbal, and implicit cue items was \< .6 (α = 0.57), while the Nonverbal (composite) and Verbal (single-item) scores were only weakly correlated (r = .12). Given the insufficient empirical support for combining all three items and the goals of the current research, this dissertation examined Nonverbal and Verbal cue use separately. Higher scores indicated greater use of each type of cue to communicate consent.

##### 2) Situational sensitivity and within-person validity

To meet this criterion, event-level scores should change in predictable ways across situations within the same person.

In Phase II, items derived and refined in Phase I were administered in a seven-day pilot study, with the ESM survey sent to participants (*n* = 12) four times per day. Following the study period, pilot participants completed an exit survey that collected feedback on the sexual consent measures and assessed their acceptability in the broader context of the ESM study design. Findings included within-person associations between internal and external sexual consent that were consistent with preexisting theory and research asserting these are separable but related facets of a multidimensional process<sup>[1](#ref-muehlenhardComplexitiesSexualConsent2016)</sup>. For instance, verbal and nonverbal consent cues were both positively associated with internal consent feelings of safety/comfort and readiness; verbal cue use was also positively associated with internal feelings of agreement/want, while nonverbal cue use was not.<sup>[51](#ref-willisDevelopingValidFeasible2021)</sup> Further, external consent communication was not associated with alcohol use at the encounter level for participants in committed relationships, replicating discriminant patterns observed in previous research.<sup>[59](#ref-jozkowskiDoesDrinkingAlcohol2015)</sup>

Recent recommendations for validating single-item ESM measures call for concurrent associations at the within-person level, as well as situational sensitivity that demonstrates predictable within-person changes linked to theory-driven contextual correlates.<sup>[55](#ref-wolfers)</sup> As a benchmark, within-person concurrent validity is present when a distinct but theoretically related variable significantly predicts the focal outcome with a standardized effect size greater than .09.<sup>[55](#ref-wolfers)</sup> For example, one ESM study using multilevel models found that a composite internal consent feelings score significantly predicted both verbal and nonverbal consent cue use and that more intimate encounters predicted higher consent communication within persons (verbal β = .53, p \< .001; nonverbal β = .45, p \< .001; situational effect β = .32, p \< .001).<sup>[10](#ref-willisAssessingWithinPersonVariability2021)</sup>.

To evaluate within-person concurrent validity in the current sample, we examined whether event-level sexual feelings (a composite of subjective arousal and pleasure; psb_feel_wp) predicted same-event verbal and nonverbal consent cue use. Results indicated that when participants reported higher-than-usual sexual feelings during a sexual encounter, they also reported using more verbal (β = 0.17, SE = 0.03, t(348) = 5.77, p \< .001, 95% CI \[0.11, 0.23\]) and nonverbal (β = 0.13, SE = 0.03, t(345) = 5.27, p \< .001, 95% CI \[0.08, 0.18\]) consent cues. These findings provide evidence of within-person concurrent validity, suggesting that higher situational sexual arousal and pleasure co-occur with greater active consent communication.

To test event-level situational sensitivity, we examined whether type of sexual behavior (Contact vs. Penetration; act_c) predicted verbal and nonverbal consent cue use. Encounters involving penetration (vs. contact-only) were associated with greater verbal cue use, β = 0.24, SE = 0.04, t(488) = 5.78, p \< .001, 95% CI \[0.16, 0.32\], and greater nonverbal cue use, β = 0.10, SE = 0.04, t(459) = 2.80, p = .005, 95% CI \[0.03, 0.17\]. These results support situational sensitivity by showing that participants’ consent communication varies appropriately with the intimacy of the sexual behavior reported.

##### 3) Between-person validity evidence

To satisfy multi-level validity, evidence should demonstrate that the measure functions both within- and between-persons. The previous section offered within-person evidence; here we present between-person evidence and then integrate across levels.

Recent research by Shi and colleagues offers support for between-person concurrent validity in a sample of Chinese adults in mixed-gender relationships.<sup>[58](#ref-shiImpactSexualConsent2025)</sup> Specifically, explicit-verbal consent communication significantly predicted greater sexual satisfaction during participants’ most recent sexual event with their partner (β = .31, SE = .01, $R^2$ = .10, *p* \< .001), and implict-nonverbal consent communication significantly predicted greater sexual satisfaction in general (β = .13, SE = .01, $R^2$ = .08, *p* \< .001).<sup>[58](#ref-shiImpactSexualConsent2025)</sup> These findings indicate that individuals who report higher levels of external consent communication also report higher sexual satisfaction, consistent with theoretical expectations.

To evaluate between-person convergent validity in the current sample, participants’ baseline reports of verbal (`v_on`: “I always ask verbally for consent at each step of a sexual encounter”) and nonverbal (`non_vrb_bsl`: “Typically, I communicate sexual consent to my partner using nonverbal signals and body language”) consent were correlated with person-level aggregates of the same behaviors from the diary (`verbal_means`: mean verbal cue use across events; `mean_nonverb`: mean nonverbal cue use across events). Aggregating diary data between persons aligned with the level of analysis for the baseline measures and respected the hierarchical structure.

**Verbal cues.** Participants who reported greater use of verbal consent cues at baseline also tended to report higher average verbal cue use across diary-reported encounters, *r*(152) = .54, *p* \< .001.

**Nonverbal cues.** The corresponding association for nonverbal consent cues was smaller but significant, *r*(152) = .20, *p* \< .001, indicating a consistent relationship between baseline and event-level nonverbal consent communication.

## Analytic Strategy

Analyses were conducted in parallel frameworks. We selected Linear Mixed-Effects Models (LMEM) for primary analyses and fit these multilevel models using the `lme4` package in R.<sup>[60](#ref-bates2015)</sup>

Cumulative Link Mixed Models (CLMMs) were selected for use a secondary robustness check. These secondary analyses were conducted in R using the `ordinal` package.<sup>[62](#ref-christensen2019)</sup> CLMMs estimate the cumulative probability of being at or above each ordered response category (i.e., level of the outcome variable) and allow for inclusion of random effects to account for the nonindependence of repeated observations. This approach also handles non-normal outcomes appropriately without assuming equal intervals between categories.<sup>[63](#ref-christensen2018)</sup>

Robustness was evaluated based on consistency in the direction and significance of effects across modeling frameworks.

### Data structure and ICCs

Each participant received one diary survey per day for 28 consecutive days. Responses were not statistically independent because participants contributed multiple event-level observations. Daily diary reports that captured sexual activity involving physical contact with another person (hereafter referred to as sexual events) were retained for analysis. The full study period spanned approximately 80 days, during which different participants sometimes reported sexual events on the same calendar date; however, no individual participant contributed more than one event for any given day. Consequently, Level 1 units were sexual-event days nested within participants (Level 2). Models included random intercepts for participant `(1|nid)` to model between-person differences in baseline cue-use propensity. Random slopes for time-varying predictors were included when supported by identifiable variance. Preliminary LMEMs were fit with and without random slopes for within-person predictors to determine the appropriate random-effects structure. Models were compared via likelihood-ratio tests (LRTs) and information criteria; those with relatively smaller values for the Akaike Index Criterion (AIC) and Bayesian Index Criterion (BIC) were considered to fit the data better.

To assess the extent to which participants’ reports of consent cue use varied within versus between individuals across sexual encounters, intraclass correlation coefficients (ICCs) were calculated for LMEMs on the transformed outcomes and via unconditional random-intercept CLMMs for ordinal outcomes. ICCs quantified the proportion of total variance in each outcome that was attributable to stable, between-person differences (range = 0–1); the remaining proportion reflected within-person variability across repeated observations. Although ICCs are conceptually identical across modeling frameworks, they must be estimated on the latent logistic scale for ordinal outcomes analyzed with CLMMs rather than on the continuous scale used in LMEMs. This distinction is important because ordinal outcomes do not have a directly estimated residual variance; instead, the residual variance is fixed at $\pi^2 / 3$ on the logistic scale.<sup>[64](#ref-nakagawa2017)</sup> The variance of the random intercept ($\sigma^2_{u0}$) was therefore extracted from each unconditional CLMM and combined with the fixed residual variance on the latent logistic scale according to the formula presented below.

$$ICC = \frac{\sigma^2_{u0}}{\sigma^2_{u0} + \pi^2 / 3}$$

### Variable preparation and centering

Time-varying predictors were decomposed into within-person and between-person components. Specifically, each time-varying predictor $Xij$ (event $i$ for person $j$) was person-mean centered to compute the person-mean deviation $X_{ij}^{c} = X_{ij} - \bar{X}_{.j}$, modeled at Level 1. Person means $\bar{X}_{.j} = \frac{1}{n_j} \sum_i X_{ij}$ were then grand-mean centered for inclusion in models as Level-2 predictors. Level-1 categorical variables also were decomposed.

Models then included both the within-person effect $X_{ij}^{c}$ (does a participant use more consent cues on days when they are above their own typical level of $X$?) and the between-person effect $\bar{X}_{.j}$ (do participants with higher typical levels of $X$ use more consent cues overall?), which yielded separate coefficients. For CLMMs, these coefficients represent distinct effects on the cumulative log-odds of endorsing higher cue-use categories.

Time-invariant, person-level variables also were grand-mean centered to place the model intercept at a meaningful reference and to reduce collinearity between intercepts and slopes. Continuous predictors were scaled per SD for interpretability. Finally, we dummy-coded nominal predictors as factors with a theoretically-neutral referent and contrast-coded ordered predictors.

In a CLMM, each predictor is assumed to have a constant effect on the cumulative log-odds of endorsing higher outcome categories across all response thresholds. Violations occur when a predictor’s influence differs for lower versus higher category transitions (e.g., between “Not at all → A little” versus “Quite a bit → Very much”). We evaluated this proportional-odds assumption using complementary statistical and visual diagnostic approaches.

First, we conducted a global `nominal_test()` of the proportional-odds model. This assumption concerns fixed effects only, so we tested it using a cumulative link model (CLM) without random effects. When a predictor was flagged during the global screen, we then fit threshold-specific binary logistic mixed models (GLMMs) to examine whether fixed-effect slopes varied substantially across cutpoints (e.g., “Not at all → A little,” “A little → Somewhat,” etc.).

Specifically, for average physical intimacy (`act_pm_c`), the estimated log-odds were 1.65, 2.07, 1.73, and 1.73 across successive thresholds, all p \< .05. Because these coefficients were consistent in direction and magnitude across thresholds (differences ≤ 0.4), the proportional-odds assumption was considered tenable. A visual inspection provided further support for this conclusion.

### Model specification

Each daily survey was distributed between 11:00 AM and 2:00 PM (with a reminder two hours later) and asked about experiences *since the previous survey*, yielding roughly 24-hour retrospective reports. Although event spacing was irregular, the daily diary design may have introduced predictable temporal structure (e.g., due to gradual study fatigue). To assess whether design-related patterns could bias fixed-effect estimates, we evaluated whether day on study accounted for systematic variation in verbal cue use. A likelihood-ratio test comparing models with and without study day as a fixed predictor indicated that including study day did not significantly improve model fit, $\chi^2(1)= 3.12$, $p = .08$.

To evaluate whether temporal trends varied across participants, we compared models with and without a random slope for study day. Although the average effect of study day on verbal consent cue use was not significant, likelihood-ratio tests indicated significant between-person variability in day-to-day slopes. Allowing the effect of study day to vary across individuals significantly improved model fit, $\chi^2(2)= 14.32$, $p < .001$. Accordingly, a random slope for study day was retained in all models predicting Yeo-Johnson transformed verbal cue use to account for individual differences in temporal trajectories.

To examine predictors of verbal and nonverbal consent cue use while accounting for both within- and between-person variability, we fit conditional mixed-effects models estimating cue use from person- and encounter-level characteristics, as specified in each hypothesis set.

#### Addressing convergence problems

Convergence problems occur when the estimation routine fails to identify a stable solution for model parameters. We based our approach to addressing these problems on Brauer and Curtin’s (2018) guidelines regarding sequential corrective steps that researchers can take when LMEM convergence fails. Their recommendations prioritize nonintrusive steps that preserve the intended model specification before moving on to classic or corrective remedies that alter the model structure.<sup>[66](#ref-brauerLinearMixedeffectsModels2018)</sup>.

Specifically, we used the full, cleaned dataset and retained all available cases following multiple imputation and data quality verification in order to include as many sexual events and participants as possible. We took steps to minimize potential convergence issues prior to model estimation by carefully screening the data for distributional irregularities and violations of model assumptions, then applying Yeo-Johnson transformations to both skewed continuous outcomes. In addition, all predictors were centered, with time-varying predictors person-mean centered to reduce multicollinearity in the random-effects structure. When convergence problems nevertheless emerged, we evaluated models for potential misspecification, ensuring that random slopes were only specified for predictors that exhibited within-person variation. All predictors and covariates were modeled as fixed effects, and a participant random intercept was included in all models to account for nonindependence.

When convergence difficulties persisted, we proceeded through the following steps in order. Specifically, models were refit using the *bobyqa* optimization algorithm, and the maximum number of optimizer iterations was increased (up to 200,000 function evaluations) to allow adequate exploration of the likelihood surface. We also evaluated the data for extremely sparse categories and units with very few observations in key design cells, then addressed these issues by collapsing or removing problematic units as needed. Finally, we were prepared to rescale predictors and, if necessary, the outcomes to improve numerical stability, but all convergence problems were resolved before advancing to this step was needed.

### Alcohol and consent cue use

**Hypothesis 1a** predicted that alcohol consumption (indexed by number of drinks) would be negatively associated with both verbal and nonverbal consent cue use. To test this hypothesis, LMEMs were fit separately for verbal and nonverbal consent outcomes, decomposing alcohol consumption into within-person (event-level) and between-person (person-mean) components and including random intercepts for participants. LMEM results partially supported H1a for verbal consent. There was a significant, negative between-person effect of alcohol consumption on verbal cue use, such that participants who generally drank more than others before sexual events reported lower verbal cue use overall ($b$ = -0.42, $SE$ = 0.13, $t$ = -3.14, $p$ = 0.002). In contrast, the within-person effect of alcohol consumption on verbal cue use was not significant ($b$ = -0.12, $SE$ = 0.08, $t$ = -1.36, $p$ = 0.175).

To evaluate whether these conclusions were robust to the ordinal measurement of verbal consent cue use, we also fit a CLMM predicting ordinal verbal cue use from the same within- and between-person components of alcohol consumption, with a random intercept for participant (logit link). Consistent with LMEM results, the within-person effect of alcohol consumption was not significantly associated with verbal cue use ($b$ = -0.09, $SE$ = 0.11, $z$ = -0.87, $p$ = 0.383). However, also consistent with LMEM results, the between-person effect was negative and statistically significant ($b$ = -0.55, $SE$ = 0.17, $z$ = -3.17, $p$ = 0.002). This pattern of findings might suggest that participants who tend to drink more before sexual encounters typically have lower odds of using higher levels of verbal consent across events.

H1a was not supported for nonverbal consent. LMEM results indicated that neither within— ($b$ = -0.12, $SE$ = 0.08, $t$ = -1.36, $p$ = 0.175) nor between— ($b$ = -0.70, $SE$ = 0.43, $t$ = -1.63, $p$ = 0.105) person effects of alcohol consumption on nonverbal cues were significant. The same pattern was observed among CLMM results for both within— ($b$ = 0.12, $SE$ = 0.10, $z$ = 1.24, $p$ = 0.215) and between— ($b$ = -0.33, $SE$ = 0.19, $z$ = -1.77, $p$ = 0.076) person effects of alcohol consumption on nonverbal cues.

**Hypothesis 1b** predicted that greater intoxication disparity—defined as the absolute difference between participants’ subjective intoxication and their perception of their partner’s intoxication—would be negatively associated with verbal consent cue use. LMEM results partially supported this hypothesis. Between-person differences in intoxication disparity were significantly associated with lower verbal consent cue use ($b$ = -0.09, $SE$ = 0.02, $t$ = -3.47, $p$ = \< .001). However, the within-person effect of intoxication disparity was not significant ($b$ = 0.00, $SE$ = 0.01, $t$ = 0.01, $p$ = 0.991).

The secondary robustness check via CLMM yielded consistent results: there was a significant, negative between-person effect of intoxication disparity on verbal cue use ($b$ = -0.11, $SE$ = 0.03, $z$ = -3.42, $p$ = \< .001), but the within-person effect of intoxication disparity was not significant ($b$ = 0.00, $SE$ = 0.01, $z$ = 0.14, $p$ = 0.887). This pattern of findings might suggest that individuals who routinely find themselves in sexual situations with mismatched intoxication levels tend to use less verbal consent in general.

**Hypothesis 1c** predicted that attendance at bars or parties would be negatively associated with both verbal and nonverbal consent cue use, with stronger effects expected for nonverbal cues. LMEM results partially supported this hypothesis. Between-person differences in same-day drinking venue attendance were significantly negatively associated with cue use during sexual encounters, and this was true for both nonverbal ($b$ = -3.69, $SE$ = 1.39, $t$ = -2.65, $p$ = 0.009) and verbal ($b$ = -1.38, $SE$ = 0.44, $t$ = -3.12, $p$ = 0.002) cues. The within-person effects of drinking venue attendance were not significant for nonverbal ($b$ = -0.20, $SE$ = 0.63, $t$ = -0.32, $p$ = 0.748) or verbal ($b$ = -0.13, $SE$ = 0.25, $t$ = -0.54, $p$ = 0.590) consent.

Similarly, across CLMMs, between-person differences in bar-or-party attendance on sexual encounter days were significantly, negatively linked to both nonverbal ($b$ = -1.77, $SE$ = 0.00, $z$ = -523.70, $p$ = \< .001) and verbal ($b$ = -1.87, $SE$ = 0.60, $z$ = -3.13, $p$ = 0.002) cue use, but no significant effects were observed for within-person deviations predicting nonverbal ($b$ = -0.21, $SE$ = 0.29, $z$ = -0.72, $p$ = 0.473) or verbal ($b$ = -0.30, $SE$ = 0.32, $z$ = -0.93, $p$ = 0.352) consent.

### Modifiers of alcohol-consent associations

**Hypothesis 2a** predicted that the negative association between alcohol consumption and verbal consent cue use would be weaker for penetrative behaviors and stronger for encounters involving kissing or touching. To test this hypothesis, interaction terms between alcohol consumption and sexual behavior type were included in mixed-effects models predicting verbal consent cue use. Contrary to Hypothesis 2a, the interaction between alcohol consumption and behavior type was not statistically significant at either the within- ($b$ = 0.11, $SE$ = 0.17, $t$ = 0.62, $p$ = 0.535) or between- ($b$ = 0.11, $SE$ = 0.21, $t$ = 0.51, $p$ = 0.612) person level, suggesting the association between pre-encounter number of drinks consumed and verbal consent cue use did not differ by behavior type in this sample.

The secondary robustness check via CLMM yielded consistent results: sexual behavior type did not significantly moderate the within- ($b$ = 0.19, $SE$ = 0.25, $z$ = 0.73, $p$ = 0.463) nor between- ($b$ = 0.02, $SE$ = 0.29, $z$ = 0.06, $p$ = 0.953) person effects of alcohol consumption on ordinal verbal cue use, suggesting the association between pre-encounter number of drinks consumed and verbal consent cue use did not differ by degree of physical intimacy (contact- v. penetrative- sexual behavior type) in this sample.

**Hypothesis 2b** predicted that the negative association between drinking venue attendance and nonverbal consent cue use would be stronger among less familiar partners. Mixed-effects models testing interactions between bar-and-party attendance and partner familiarity revealed no statistically significant moderation effects at either the within- or between-person level ($ps$ \> .219). Thus, the association between drinking venue attendance and nonverbal cue use did not differ as a function of partner familiarity in this sample.

**Hypothesis 2c** predicted that the negative association between intoxication disparity and verbal consent cue use would be stronger among men compared to women and gender-nonconforming individuals. Mixed-effects models were consistent in that neither the LMEM ($b$ = -0.01, $SE$ = 0.02, $t$ = -0.32, $p$ = 0.749) nor the CLMM ($b$ = -0.01, $SE$ = 0.02, $z$ = -0.34, $p$ = 0.736) results supported this hypothesis. Interaction terms between intoxication disparity and gender were not statistically significant, indicating that the the within-person slope of intoxication disparity did not differ by gender in this sample.

### Distal individual factors

Consistent with **Hypothesis 3a**, LMEM results revealed that social desirability bias ($b$ = 0.46, $SE$ = 0.18, $t$ = 2.50, $p$ = 0.013) and rape myth rejection ($b$ = 0.36, $SE$ = 0.17, $t$ = 2.15, $p$ = 0.033) were significantly, positively associated with verbal consent cue use after controlling for partner familiarity and behavior type. Results of the secondary robustness check also supported H3a, such that CLMM findings indicated that social desirability bias ($b$ = 0.65, $SE$ = 0.25, $z$ = 2.56, $p$ = 0.011) and rape myth rejection  
($b$ = 0.52, $SE$ = 0.24, $z$ = 2.21, $p$ = 0.027) positively predicted endorsement of verbal cues.

**Hypothesis 3b** predicted that permissive individual and perceived peer attitudes toward alcohol and coercion will be negatively associated with verbal cues after controlling for partner familiarity and behavior type. Results partially supported this hypothesis. Perceived peer attitudes were significantly, negatively associated with verbal cue use ($b$ = -0.49, $SE$ = 0.22, $t$ = -2.27, $p$ = 0.025), whereas permissive *individual* attitudes were not significantly associated with verbal cues when controlling for all other predictors in the model ($b$ = -0.24, $SE$ = 0.23, $t$ = -1.03, $p$ = 0.307). CLMM results yielded the same pattern of findings: after controlling for familiarity and sexual behavior, perceived peer attitudes ($b$ = -0.62, $SE$ = 0.31, $z$ = -2.02, $p$ = 0.043) significantly predicted verbal consent cues while personal attitudes ($b$ = -0.41, $SE$ = 0.32, $z$ = -1.30, $p$ = 0.194) did not.

**Hypothesis 3c** predicted that hazardous alcohol consumption would be negatively associated with both verbal and nonverbal consent cue use after controlling for partner familiarity and behavior type. H3c was not supported for either outcome or within either analysis framework. Across both verbal ($b$ = -0.30, $SE$ = 0.18, $t$ = -1.64, $p$ = 0.103) and nonverbal ($b$ = -0.60, $SE$ = 0.59, $t$ = -1.01, $p$ = 0.314) cues, trait-level hazardous drinking did not uniquely predict consent communication when the LMEM included contextual characteristics of the sexual event as covariates. CLMM results yielded the same pattern for both verbal ($b$ = -0.50, $SE$ = 0.26, $z$ = -1.92, $p$ = 0.055) and nonverbal ($b$ = -0.35, $SE$ = 0.26, $z$ = -1.34, $p$ = 0.181) cues.

**Hypothesis 3d** predicted that permissive individual attitudes toward alcohol and coercion would strengthen the negative association between intoxication disparity and verbal cue use after controlling for partner familiarity and behavior type. To test this hypothesis, interaction terms between intoxication disparity and permissive attitudes were included in mixed-effects models alongside partner familiarity and behavior type. Contrary to Hypothesis 3d, the interaction between disparity and permissive attitudes was not statistically significant at the within- ($b$ = -0.00, $SE$ = 0.01, $t$ = -0.24, $p$ = 0.814) or between- ($b$ = -0.01, $SE$ = 0.03, $t$ = -0.47, $p$ = 0.639) person level, indicating that the negative association between mismatched intoxication and verbal consent cue use was not modified by personal attitudes.

CLMM results were consistent with those derived from the primary LMEM. The within-person slope of intoxication disparity did not differ by permissive attitudes ($b$ = -0.01, $SE$ = 0.01, $z$ = -0.44, $p$ = 0.659), and the between-person effect of intoxication disparity on verbal cues did not vary as a function of individual attitudes toward alcohol and coercion ($b$ = -0.02, $SE$ = 0.04, $z$ = -0.40, $p$ = 0.688).

## Limitations

The proposed research is not without limitations. First, there were measures not included in this data collection effort that we intend to prioritize in the future. Considering this project focuses on active consent communication, we did not measure passive “no response” signals, which are commonly reported cues used to infer and communicate consent among young adults despite risks associated with their ambiguity and inconsistent associations with internal willingness. Also, we did not assess refusal cues that participants may have used to indicate lack of willingness, so it is possible that participants were also reflecting on behaviors they used to assert boundaries when responding to items assessing verbal and nonverbal consent cues. Other measures not included that could have strengthened the current study are perceptions of partner cue usage, because we cannot speak to how participants inferred consent from their partners without those data, nor can we know whether the body language, behaviors, or comments that participants interpreted as cues were intended to signal consent. In future research, these limitations can be addressed through dyadic data collection, though recruiting in pairs introduces a different set of constraints with respect to feasibility and generalizability. Finally, we did not provide options to select from or ask participants to report specific examples of the behaviors that they were classifying as verbal or nonverbal cues, and we did not provide participants with the operational definitions for these constructs prior to study initiation. 

Specific characteristics of the current study sample and design also may limit what can be accomplished with the proposed research. Participants were recruited via Sona Systems and compensated with extra credit in their introductory psychology class, resulting in a college-student sample with 85% of participants in their first year. More than 90% of the sample fell below age 20, while a sample with more proportional representation of upperclassmen may have yielded a greater number of partnered sexual events and drinking episodes for analysis. Further, as these data were collected from college students at a primarily white institution in the midwestern U.S., the sample is homogeneous in terms of race (77% White) and sexual orientation (87% heterosexual). Directions for future research include more stringent criteria that require eligible study participants be sexually active; prior sexual experience was not an inclusion criterion for participating in the current research, as it is common for young adults’ first partnered sexual experiences to occur during their first semester at university.<sup>[67](#ref-follingstad2023)</sup> 

With respect to study design, the experience sampling method used to support this project was a relative strength compared to the primarily cross-sectional literature in this area. Using electronic diaries reduced the timeframe for retrospective recall, however, survey timing and framing still required participants to think backwards when reporting on behaviors and experiences that preceded sexual events (e.g., pre-encounter number of drinks) on the previous day. Thus, results examining event-level variables can only be interpreted as associational and predictive within the concurrent window. Additionally, asking participants to rate, for example, a sexual partner’s intoxication level relative to their own at the same timepoint that we asked them about sexual consent communication may have increased susceptibility to social desirability bias, potentially leading participants to respond in ways they thought would be perceived more positively by the research team, or in ways that reduced their own cognitive dissonance around possibly taking advantage of an intoxicated person or being taken advantage of themselves. In future research, these limitations could be addressed using ecological momentary assessment with ≥2 surveys per day or event-contingent sampling, permitting assessment of contextual variables pre-encounter and concurrent or post-encounter variables immediately thereafter. Finally, there is risk of overfitting to sample-specific nuance, as we are fitting and evaluating our models in the same sample; techniques for estimating out-of-sample generalizability were considered but ultimately omitted from the analytic plan, which prioritizes feasibility. Directions for future research include goals to replicate and extend this research in a separate college sample of new participants, as well as among adults in other sociocultural contexts.
