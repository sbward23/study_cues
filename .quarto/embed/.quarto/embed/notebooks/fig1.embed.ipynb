```python
library(tidyverse)
```

```python
# Create data
data <- tibble(
  name = c("A","B","C","D","E") ,  
  value = c(3,12,5,18,45)
  )
```

```python
#| label: fig-1
#| fig-cap: "A Basic Barplot Figure"

ggplot(data, aes(x=name, y=value)) + 
  geom_bar(stat = "identity")
```
