```
library(tidyverse)
```

```
knitr::opts_chunk$set(dev = "png")
```

```
icc_table <- readRDS(here::here("objects/icc_table.rds"))
```

```

icc_table <- icc_table %>%
  mutate(
    ICC = ICC |> str_replace("^0\\.", "."),
    `1 - ICC` = `1 - ICC` |> str_replace("^0\\.", "."),
    `95% CI` = `95% CI` |> str_replace("\\b0\\.", ".")
  )

icc_table %>%
  kbl(
    booktabs = TRUE,
    align = c("l", "c", "c", "c")
  ) %>%
  kable_styling(full_width = FALSE)
```
